```python
# QUESTION.3 You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the
#  DataFrame and calculates the sum of the first three values in the 'Values' column. Thefunction should print the sum to
#  the console.
#  For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should calculate and
#  print the sum of the first three values, which is 60.

# ANSWER 
data = {
    'Values':[10,20,30,40,50,60]
}
```

```python
df = pd.DataFrame(data)
```

```python
def sum_first_three_values(df):
    
    total_sum = 0
    
    
    for index, row in df.iterrows():
      
        if index < 3:
            
            total_sum += row['Values']
        else:
            
            break
    
    
    print("Sum of the first three values:", total_sum)
```

```python
sum_first_three_values(df)
```

```python
# QUESTION.4 Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column
# 'Word_Count' that contains the number of words in each row of the 'Text' column.

import pandas as pd

def add_word_count_column(df, text_column_name='Text', word_count_column_name='Word_Count'):
    
    df[word_count_column_name] = df[text_column_name].apply(lambda x: len(str(x).split()))
    return df

# Example usage:
# Create a sample DataFrame
data = {'Text': ["This is a sample text", "Count the words here", "Pandas is awesome"]}
df = pd.DataFrame(data)


df = add_word_count_column(df)


print(df)
```

```python
# QUESTION.7 You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in
# the format 'username@domain.com'. Write a Python function that creates a new column'Username' in df that contains
# only the username part of each email address.

# ANSWER 
def extract_username_from_email(df):
    
    df['Username'] = df['Email'].str.split('@').str[0]
    return df


data = {'Email': ['user1@example.com', 'user2@example.com', 'user3@example.com']}
df = pd.DataFrame(data)
df = extract_username_from_email(df)
print(df)
```

```python
# QUESTION.9 Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean,
# median, and standard deviation of the values in the 'Values' column.
# ANSWER 
data = {
    'Values':[10,20,30,40,50]
}
```

```python
df = pd.DataFrame(data)
```

```python
def calculate_statistics(df):
    
    mean = df['Values'].mean()
    median = df['Values'].median()
    std_deviation = df['Values'].std()
    
    return mean, median, std_deviation
```

```python
calculate_statistics(df)
```

```python
# QUESTION.10 Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create
# a new column 'MovingAverage' that contains the moving average of the sales for the past 7 daysfor each row in the
# DataFrame. The moving average should be calculated using a window of size 7 andshould include the current day.
# ANSWER 
DATA = {'Date': ['2023-09-01', '2023-09-02', '2023-09-03', '2023-09-04', '2023-09-05'],
                   'Sales': [10, 15, 20, 25, 30]}
```

```python
df = pd.DataFrame(DATA)
```

```python
df
```

```python
def calculate_moving_average(df):
    # Sort the DataFrame by the 'Date' column if it's not already sorted
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values(by='Date')

   
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()

    return df
```

```python
df = calculate_moving_average(df)
```

```python
df
```

```python
# QUESTION.11 You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column
# 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g. Monday, Tuesday) corresponding
#  to each date in the 'Date' column.
# For example, if df contains the following values:
# Date
# 2023-01-01
# 2023-01-02
# 2023-01-03
# 2023-01-04
# 2023-01-05

# ANSWER 
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
def add_weekday_column(df, date_column_name='Date', new_column_name='Weekday'):
    
    df[date_column_name] = pd.to_datetime(df[date_column_name])

    df[new_column_name] = df[date_column_name].dt.strftime('%A')
```

```python
add_weekday_column(df)
```

```python
print(df)
```

```python
# QUESTION.12 Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to
# select all rows where the date is between '2023-01-01' and '2023-01-31'.

# ANSWER 
start_date = '2023-01-01'
end_date = '2023-01-31'

def select_rows_in_date_range(df, start_date, end_date):
   
    df['Date'] = pd.to_datetime(df['Date'])
    
    
    mask = (df['Date'] >= start_date) & (df['Date'] <= end_date)
    
    
    selected_rows = df[mask]
    
    return selected_rows
```
